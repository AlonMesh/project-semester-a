# Spliting

```python
#spliting the data to 2:
prices = cars['price'] #only prices
carsWithoutPrice = cars.copy().drop(columns = "price") #all but prices
```

```python
#creating train and test
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(carsWithoutPrice, prices , test_size=0.1 , random_state = 42)
```
