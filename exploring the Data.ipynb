```python
car_names = cars.CompanyName.unique()
corol_com = np.zeros(len(car_names))
counter = 0

for x in car_names:
    tempStr = "is_" + itemToEnglish(x)
    corol_com[counter] = abs(corr_matrix["price"].sort_values(ascending = False)[tempStr])
    counter = counter+1;
```

```python
plt.figure(figsize=(25, 6))
plt.bar(car_names, corol_com, width=0.5)
plt.xlabel('Company')
plt.ylabel('Correlation to price')
plt.title('How much each company infulence the price?')
```
