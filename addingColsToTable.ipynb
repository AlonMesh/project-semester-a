```python
#cars['annual_licensing_fee'].astype(int):

def cov(value):
    if type(value) != int:
        value = value.replace('\u200f', '').replace('\xa0₪', '').replace(',', '')
        return int(value)
    return value
    
cars['annual_licensing_fee'] = cars['annual_licensing_fee'].apply(cov)

cars.head(5)
```
